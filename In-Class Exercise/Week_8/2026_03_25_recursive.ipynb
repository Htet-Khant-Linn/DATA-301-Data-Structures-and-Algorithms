# Recursive

## Recursive Method

```python
def recursiveMethod(n):
  if n < 1:
    print("n value is less than 1.")
  else:
    recursiveMethod(n-1)
    print(n)

recursiveMethod(4)
```

## Power of Two

```python
def powerofTwo(n):
  if n==0:
    return 1
  else:
    power = powerofTwo(n-1)
    return power * 2

print(powerofTwo(4))
```

```python
def powerofTwo(n):
  i = 0
  power = 1
  while i < n:
    power = power * 2
    i = i + 1
  return power


print(powerofTwo(4))
```

## Count Down

### Using For Loop

```python
def countdown(n):
  for i in range(n):
    print(n)
    n = n - 1
  print("Count down completed.")
countdown(5)
```

### Using For Loop

```python
def countdown(n):
  for i in range(n, 0, -1):
    print(i)
  print("Count down completed.")

countdown(5)
```

### Using recursive approach

```python
def countdown(n):
  if n < 1:
    print("Count down completed.")
  else:
    print(n)
    countdown(n-1)


countdown(5)
```

```python
def countdown(n):
  if n == 0:
    return
  else:
    print(n)
    countdown(n-1)

countdown(5)
```

## Reverse String

### For Loop Approach

```python
def reverse_string(s):
  reversed_string = ''
  for char in s:
    print(char)
    reversed_string = char + reversed_string
  print(reversed_string)

reverse_string("Hello")
```

```python
def reverse_string(s):
  if s == "":
    return ""
  else:
    return reverse_string(s[1:]) + s[0]

print(reverse_string("May Mon Thant"))
```

## Factorial

```python
def factorial(n):
  assert n >= 0 and int(n) == n, 'The number must be positive integer only'
  if n in [0, 1]:
    return 1
  else:
    return n * factorial(n - 1)

print(factorial(3))
```

```python
import sys    # set the recursion limit
sys.setrecursionlimit(50)

def factorial(n):
  print(n)
  return n * factorial(n-1)

factorial(5)
```

## Fibonacci Numbers

```python
def faibonacci(n):
  assert int(n) == n and n >= 0, "Fibonacci number cannot be negative or non integer."
  if n in [0, 1]:
    return n
  else:
    return faibonacci(n-1) + faibonacci(n-2)

print(faibonacci(6))
```

Time complexity:
- factorial -> O(n)
- fibonacci -> O(2^n)
