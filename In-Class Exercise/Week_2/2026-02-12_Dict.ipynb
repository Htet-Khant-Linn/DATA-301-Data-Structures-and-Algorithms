## 1. Create Dict

```python
# Creating Dictionary

my_dict = dict()
print(my_dict)
```

```python
my_dict2 = {}
print(my_dict2)
```

```python
eng_number = dict(one = '1', two = '2', three = '3')
print(eng_number)
```

```python
eng_number2 = {'one':'1', 'two':'2', 'three':'3'}
print(eng_number2)
```

```python
eng_number_list = [('one', '1'), ('two', '2'), ('three','3')]
eng_number3 = dict(eng_number_list)
print(eng_number3)
```

## Insertion to Dict
