## 3. Traverse Dict

```python
# need to check
def traverse_dict(dict):
    for key in dict:
        print(f"Key: {key}, Value: {dict[key]}")

traverse_dict(my_dict2)
```

## 4. Search Through Dictionary

```python
# add to the dict
my_dict2['major'] = 'SDS'
print(my_dict2)
print()


def search_dict(dict, value):
    for key in dict:
        if dict[key] == value:
            return key, value
    return 'The values does not exist!'


print(search_dict(my_dict2, 'SDS'))

```

## 8. Dict Comprehension

```python
import random
random.seed(823)
```

```python
city_names = ['Paris', 'London', 'Rome', 'Berlin', 'Madrid']
city_temps = {city:random.randint(20,30) for city in city_names}
print(city_temps)
```

```python
for key in city_temps:
    print(city_temps[key])
```

```python
above_25 = [city for city, temp in city_temps.items() if temp > 25]         # generate list
print(above_25)

above_25v2 = {city:temp for city, temp in city_temps.items() if temp > 25}  # generate dict
print(above_25v2)
```

```python
below_25 = [city for city, temp in city_temps.items() if temp < 25]         # generate list
print(below_25)

below_25v2 = {city:temp for city, temp in city_temps.items() if temp < 25}  # generate dict
print(below_25v2)
```
