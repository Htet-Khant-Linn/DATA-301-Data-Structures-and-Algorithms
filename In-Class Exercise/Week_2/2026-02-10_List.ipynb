## 2. Traverse List

```python
shopping_list = ['Milk', 'Strawberry', 'Chocolate', 'Toy']
print('-------')
print(shopping_list[2])
print('-------')
print('Milk' in shopping_list) # checking wheter the item is in the list
print('-------')

def traverse_list(list):
    for i in shopping_list:
        print(i)

traverse_list(shopping_list)
print('-------')

def traverse_list_2(list):
    for i in range(len(list)):
        list[i] = list[i] + " - "
        print(list[i])
traverse_list_2(shopping_list)
```

## 3. Insert List

```python
myList = [1,2,3,4,5,6,7]
print(myList)

myList[2] = 33
print(myList)

myList[4] = 55
print(myList)

myList.insert(4, 15)
print(myList)

newList = [11, 12, 13, 14]
myList.append(newList)      # append will add a nested list
print(myList)

myList = [1,2,3,4,5,6,7]
newList = [11, 12, 13, 14]
myList.extend(newList)      # extend will add without creating a nested list
print(myList)
```

## 4. Delete to List

```python
myList2 = ['a', 'b', 'c', 'd', 'e', 'f']
print(myList2)
print()

myList2[0:2] = ['x', 'y']
print(myList2)
print()

print('--- pop() ---')
myList2.pop(1)      # use index number
print(myList2)
print()

print('--- del ---')
del myList2[0:2]    # use slicing
print(myList2)
print()

print('--- remove() ---')
myList2.remove('d') # use str
print(myList2)
```

## 5. Search Element List

```python
myList3 = ['a', 'b', 'c', 'd', 'e', 'f']

def linear_search(list, target):
    for i, value in enumerate(list):
        if value == target:
            return i
    return "Target not found"

print(f'The index location is: {linear_search(myList3, 'e')}.')
print(f'The index location is: {linear_search(myList3, 'g')}.')
```

## 6. List Operations

```python
myList4 = list()

while True:
    input_num = input("Enter a number: ")
    if input_num == 'done':
        break
    value = float(input_num)
    print(f'You entered: {value}.')
    myList4.append(value)

average_num = sum(myList4) / len(myList4)

print(f'Avarage of list {myList4} is {average_num}.')
```

## 11. Conditional List Comprehension

```python
myList7 = my_List = [-1, 10, -20, 2, -90, 60, 45, 20]

new_list2 = [number for number in myList7 if number > 10]
print(new_list2)
```

```python
new_list3 = [number ** 2 for number in myList7 if number > 0]
print(new_list3)
```

```python
new_list4 = [number if number > 0 else 0 for number in myList7]
print(new_list4)
```

```python
def get_number(number):
    if number > 0:
        return number
    else:
        return 'Negative'
    
new_list5 = [get_number(number) for number in myList7]
print(new_list5)
```

```python
sentence = "My name is Htet Khant Linn. 123."

def is_consonant(letter):
    vowels = 'aeiou'
    return letter.isalpha() and letter.lower() not in vowels

consonants = [i for i in sentence if is_consonant(i)]
print(consonants)
```
