## 2. Search Tuples

```python
new_tuple2 = ('a', 'b', 'c', 'd', 'e')
print(new_tuple2)
```

```python
# search the element in the Tuple

def search_tuple(my_tuple, target):
    for i in range(0, len(my_tuple)):
        if my_tuple[i] == target:
            return f"The element is found at {i} index."
    return "The element is not found."

print(search_tuple(new_tuple2, 'c'))
```

```python
list = ['a', 'b', 'c', 'd', 'e']
tuple = ('a', 'b', 'c', 'd', 'e')

print(len(list))
print(len(tuple))
```

## 3. Tuple Operations

```python
my_tuple1 = (1,4,3,2,5)
my_tuple2 = (1,2,6,8,8,7)

print(my_tuple1 + my_tuple2)

print(my_tuple1 * 4)

print(4 in my_tuple1)

print(my_tuple1.count(2))
print(my_tuple2.count(8))

print(my_tuple1.index(4))

print(len(my_tuple1))

print(max(my_tuple1))

# left one item
```
