# 1_FibonacciMemoization.py

```python
# Fibonacci Series with Memoization
# Top down approach

def fibMemo(n, memo):
    if n == 1:
        return 0
    if n == 2:
        return 1
    if not n in memo:
        memo[n] = fibMemo(n-1, memo) + fibMemo(n-2, memo)
    return memo[n]

myDict = {}
print(fibMemo(6, myDict))
print(myDict)
```

# 2_FibonacciTabulation.py

```python
# Fibonacci Series with Tabulation
# Bottom Up Approach

def fibTab(n):
    tb = [0, 1]
    for i in range(2, n+1):
        tb.append(tb[i-1] + tb[i-2])
    print(tb) # for checking
    return tb[n-1]

print(fibTab(6))
```

# 3_NumberFactor_TD.py

```python
# Number Factor Problem using Dynamic Programming in Python
# Top Down Approach


def numberFactorTD(n, tempDict):
    if n in (0, 1, 2):
        return 1
    if n == 3:
        return 2
    if n not in tempDict:
        subP1 = numberFactorTD(n-1, tempDict)
        subP2 = numberFactorTD(n-3, tempDict)
        subP3 = numberFactorTD(n-4, tempDict)
        tempDict[n] = subP1 + subP2 + subP3
    return tempDict[n]

tempDict = {}
print(numberFactorTD(5, tempDict))
print(tempDict)

print('=====')
print(numberFactorTD(11, tempDict))
print(tempDict)
```

# 4_NumberFactor_BU.py

```python
# Number Factor Problem using Dynamic Programming in Python
# Bottom Up Approach

def numberFactorBU(n):
    tempArr = [1, 1, 1, 2]
    for i in range(4, n+1):
        tempArr.append(tempArr[i-1] + tempArr[i-3] + tempArr[i-4])
    return tempArr[n]


print(numberFactorBU(5))
print("=====")
print(numberFactorBU(11))
```
