# Linear Search

```python
def linearSearch(list, target):
    for i in range(len(list)):
        if list[i] == target:
            return i
    return -1

print(linearSearch([20, 50, 90, 30, 40], 90))

print(linearSearch([20, 50, 90, 30, 40], 100))
```

# Binary Search

```python
import math

def binarySearch(list, target):
    start = 0
    end = len(list)-1
    middle = math.floor((start+end)/2)

    while not (list[middle]==target) and start <= end:
        if target < list[middle]:
            end = middle-1
        else:
            start = middle+1
        middle = math.floor((start+end)/2)

    if list[middle] == target:
        return middle
    
    else:
        return -1

my_array = [8, 9, 12, 15, 17, 19, 20, 21, 28]
print(binarySearch(my_array, 17))
```
