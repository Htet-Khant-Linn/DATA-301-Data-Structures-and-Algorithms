## Exercise 1: Car Class

## 2. Rectangle OOP

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
```

## 3. Bank Account OOP

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Status: Successful.")
        print(f"Deposited: {amount} to {self.owner}'s account.")
        print()

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print("Withdraw successful.")
            print(f"You withdrew {amount} from your account.")
            self.show_balance()
            print()
        else:
            print("Insufficient balance.")
            print(f"You requested {amount} to withdraw.")
            self.show_balance()
            print()

    def show_balance(self):
        print(f"Current Balance: {self.balance}.")
        print()

```

```python
account1 = BankAccount("Charles", 5000)
account1.show_balance()
account1.deposit(100000)
account1.show_balance()
account1.deposit(5000)
account1.withdraw(10000)
```

## Exercise 3: Wallet class

attributes owner and balance

methods -> add_money(amount), spend_money(amount), check_balance()

```python
class Wallet:
    def __init__(self, owner, balance = 0):
        self.owner = owner
        self.balance = balance

    def add_money(self, amount):
        self.balance += amount
        print(f"${amount:.2f} was added to your wallet.")
        print()

    def spend_money(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"You spend ${amount:.2f} from your wallet.")
            print()
        else:
            print("Not enough money in your wallet!")
            print()

    def check_balance(self):
        print(f"Current Balance is ${self.balance:.2f}.")
        print()
```

```python
wallet1 = Wallet("Charles")
wallet1.add_money(3000)
wallet1.check_balance()
wallet1.spend_money(1000)
wallet1.check_balance()
wallet1.spend_money(3000)
```

## 4. Shopping Cart OOP

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_items(self, item, price):
        self.items.append((item, price))
        print(f"{item} added to cart and its price is {price:.2f}.")

    def remove_item(self, item, price):
        for i in range(len(self.items)):
            n, p = self.items[i]
            if n == item and p == price:
                self.items.pop(i)
                print(f"{item} removed from cart.")
                return
        print("Item not found!")


    def show_cart(self):
        print("Cart items: ")
        for product in self.items:
            print(product)
```

```python
cart1 = ShoppingCart()
cart1.add_items("Apple", 10)
cart1.add_items("Banana", 5)
cart1.show_cart()
cart1.remove_item("Apple", 10)
cart1.show_cart()
```
