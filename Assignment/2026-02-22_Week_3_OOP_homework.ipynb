```python
class Student:
    def __init__(self):
        self.students = []

    def add_students(self, student_name, subject, mark):
        self.students.append((student_name, subject, mark))
        print(f"Student: {student_name} successfully added with Mark: {mark} for {subject}.")

    def remove_student(self, student_name, subject):
        for i in range(len(self.students)):
            n, s, g = self.students[i]
            if n == student_name and s == subject:
                self.students.pop(i)
                print(f"Student: {student_name} for {subject} was removed.")
                return
        print(f"Student: {student_name} for {subject} not found.")

    def calculate_avg_mark(self):
        if len(self.students) == 0:
            print("No students to calculate an average.")
            return
            
        total_mark = 0
        for i in range(len(self.students)):
            n, s, g = self.students[i]
            total_mark += g
        
        average_mark = total_mark / len(self.students)
        print(f"Average mark of {self.students[0][0]} is {average_mark}.")

    def check_student_info(self):
        print('=== Student Information ===')
        for student in self.students:
            print(student)
```

```python
student1 = Student()
student1.add_students("Htet Khant Linn", "Python", 100)
student1.add_students("Htet Khant Linn", "R", 90)
student1.add_students("Htet Khant Linn", "Probability", 95)
student1.add_students("Htet Khant Linn", "Time Series", 99)
student1.check_student_info()
```
