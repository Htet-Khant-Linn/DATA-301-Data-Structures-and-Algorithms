1. Create one dimensional array and traverse

```python
my_one_d_array = array('i', [2, 4, 6, 8, 10])

def array_traverse(array):
    for i in array:
        print(i)
print("=== Traversed Array ===")
array_traverse(my_one_d_array)
print("======")
```

2. Access individual elements through indexes

```python
def access_individual_element(array, index):
    if index > len(array):
        print(f"The input index value: {index} can't be find in the array list.")
    else:
        print(f"The number with index value [{index}] is {array[index]}.")

access_individual_element(my_one_d_array, 3)
```
