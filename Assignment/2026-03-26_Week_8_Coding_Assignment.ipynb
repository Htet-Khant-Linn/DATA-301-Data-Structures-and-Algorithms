# Factorial of n

```python
def factorial(n):
    if n < 0:   # to check negative integer
        return "The number must be positive integer only."
    
    result = 1
    for i in range(n, 1, -1):
        result = result * i
    return result

print(f"Factorial for n = 3: {factorial(3)}")
print(f"Factorial for n = 7: {factorial(7)}")
print(f"Factorial for n = -3: {factorial(-3)}")
```

# Fibonacci Number

```python
def fibonacci(n):
    if n < 0:   # to check negative integer
        return "The number must be positive integer only."
    # Handles the first two Fibonacci numbers
    elif n == 0:
        return 0
    elif n == 1:
        return 1
    
    a, b = 0, 1     # assign a, b the starting point for the calculation

    for _ in range(2, n + 1):   # start from 2 and n inclusive
        a, b = b, a+b
    return b

print(fibonacci(0))
print(fibonacci(1))
print(fibonacci(2))
print(fibonacci(6))
print(fibonacci(10))
print(fibonacci(-3))
```
